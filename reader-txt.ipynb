```python
import os
from googletrans import Translator

dir1 = r'C:\Users\joao.coaglio\Downloads\joao_latest_str_full\latest_krjp'
dir2 = r'C:\Users\joao.coaglio\Downloads\joao_latest_str_full\ours'



def translate_japanese_to_english(text):
    translator = Translator()
    try:
        translated = translator.translate(text, src='ja', dest='en')
        if translated.text:
            return translated.text
        else:
            return text  
    except Exception as e:
        print("Erro na tradução:", str(e))
        return text 

def format_object(data_array):
    formatted_text = "\n".join(data_array)
    return formatted_text


def get_code_value(filename):
    encodings = ['utf-8', 'utf-16', 'cp1252']
    result_list = []

    for encoding in encodings:
        try:
            with open(filename, 'r', encoding=encoding) as file:
                lines = file.readlines()
                array1 = []
                for idx, line in enumerate(lines):
                    if line.startswith("<cateType>"):
                        array1.append({'cate': line.strip(), 'code': lines[idx + 1].strip()})
                result_list.extend(array1)
            return result_list
        except UnicodeDecodeError:
            pass
    
    return None

def find_codes(items, filename):
    encodings = ['utf-8', 'utf-16', 'cp1252']
    for encoding in encodings:
        try:
            with open(filename, 'r', encoding=encoding) as file:
                lines = file.readlines()
                context_lines = []

                for item in items:
                    for idx, line in enumerate(lines):
                            if line.strip() == item['cate'].strip() and lines[idx + 1].strip() == item['code'].strip():
                                context_lines.append(line.strip())  # Linha acima
                                context_lines.append(lines[idx + 1].strip())  # Linha atual
                                context_lines.append(lines[idx + 2].strip())
                                context_lines.append(lines[idx + 3].strip())
                                context_lines.append(lines[idx + 4].strip())  # Linha abaixo

                
                return context_lines
        except UnicodeDecodeError:
            pass
    
    return None


for root, _, files in os.walk(dir1):
    for file in files:
        file_path1 = os.path.join(root, file)
        file_path2 = os.path.join(dir2, file)

        newer = get_code_value(file_path1)
        our = get_code_value(file_path2)

       
        set1 = {str(item) for item in newer}
        set2 = {str(item) for item in our}

        elements_only_in_set1 = set1 - set2


        result_list = [eval(item) for item in elements_only_in_set1]
       
       
        file_translate = find_codes(result_list,file_path1)
      
       
        teste = format_object(file_translate)
      
        parts = file_path1.split('\\')
        last_part = parts[-1]

        with open(last_part, 'w', encoding='utf-8') as f:
            f.write(teste)
```
